```python
import pandas as pd
df = pd.read_csv('/kaggle/input/sales-data/kaggle sale.csv')
```

# Preprocessing

```python
print(df.head())
```

```python
print(df.info())
```

```python
df['satisfied '].head()
```

```python
df = df.rename(columns = {'satisfied ':'satisfied'})
print(df.head())
```

```python
df = df.drop(columns=['User ID'])
```

```python
print(df.head())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
columns = ['Gender', 'satisfied']
for i in columns:
    df[i] = le.fit_transform(df[i])

# columns = ['Gender', 'satisfied']  # Ensure the column names match the modified names

# for col in columns:
#     if col in df.columns:  # Check if the column exists
#         df[col] = le.fit_transform(df[col])
#     else:
#         print(f"Column '{col}' not found in DataFrame.")
```

```python
print(df.head())
```

```python
x = df.iloc[:,:-1]
print(x.head())
```

```python
y = df.iloc[:,-1]
print(y.head())
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(x_test.shape)
```

# DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier(criterion='gini',
                            splitter='best',
                            random_state=42)
```

```python
dtc.fit(x_train,y_train)
```

```python
prediction = dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print("Decision Tree Classifier Accuracy:", accuracy_score(y_test, prediction))
```

```python
print("Classification Report:\n", classification_report(y_test, prediction))
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, prediction))
```

# Naive Base Classifier

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnc = GaussianNB()
```

```python
gnc.fit(x_train,y_train)
```

```python
pred = gnc.predict(x_test)
```

```python
print("Naive Bayes Classifier Accuracy:", accuracy_score(y_test, pred))
```

```python
print("Classification Report:\n", classification_report(y_test, pred))
```
